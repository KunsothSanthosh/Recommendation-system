```python
book.rename(columns={'User.ID':'user_id','Book.Title':'book_title','Book.Rating':'book_rating'},inplace = True)
```

```python
book.columns
```

```python
book1 = book.drop("Unnamed: 0",axis=1)
```

```python
book2 = book1.drop_duplicates(keep=False)
```

```python
#number of unique userid's in the dataset
len(book2.user_id.unique())
```

```python
book2 = book2.drop_duplicates('user_id')
```

```python
#number of unique movies in the dataset
len(book2.book_title.unique())
```

```python
user_book2 = book2.pivot(index='user_id',
                                 columns='book_title',
                                 values='book_rating').reset_index(drop=True)
```

```python
user_book2
```

```python
user_book2.index = book.user_id.unique()
```

```python
user_book2
```

```python
#Impute those NaNs with 0 values
user_book2.fillna(0, inplace=True)
```

```python
user_book2
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( user_book2.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
#Set the index and column names to user ids 
user_sim_df.index = book2.user_id.unique()
user_sim_df.columns = book2.user_id.unique()
```

```python
user_sim_df.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)
```
